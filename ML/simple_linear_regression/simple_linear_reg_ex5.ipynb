```python
# simple_linear_regression_example_5
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split

# loading the dataset
mydata = pd.read_csv('sample-video-view-data.csv')

# extracting the 'dependent' and 'independent' variables from the dataset
independent_variable = mydata.iloc[ : , 0:1]
# independent_variable

dependent_variable = mydata.iloc[ : , 1:2]
# dependent_variable

# splitting the data into training data and testing data
x_train, x_test, y_train, y_test = train_test_split(independent_variable, dependent_variable, test_size=1/5, random_state=0)

# checking the distribution of the data
plt.scatter(x_train, y_train, color='magenta')
plt.xlabel('Days online(x_train)')
plt.ylabel('Number of views(y_train)')
plt.suptitle('Training_Dataset')

# creating the linear_regression_object
myregobj = linear_model.LinearRegression()
myregobj

# training the model using the training data
myregobj.fit(x_train, y_train)

# drawing the regression line
plt.plot(x_train, myregobj.predict(x_train), color='yellow')
myregobj.predict([[20]])
```

```python
# testing the model using the testing data
myregobj.fit(x_test, y_test)
plt.scatter(x_test, y_test, color='tomato')
plt.xlabel('Days online(x_test)')
plt.ylabel('Number of views(y_test)')
plt.suptitle('Testing_Dataset')
plt.plot(x_test, myregobj.predict(x_test), color='g')
myregobj.predict([[76]])
```

```python
# checking the score of the model

myregobj.score(x_test, y_test)
myregobj.score(x_train, y_train)
```
