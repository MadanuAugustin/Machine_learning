```python
# simple_linear_regression_example_4
# importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split

# loading the dataset
mydataset = pd.read_csv('Salary_Data.csv')
mydataset

# extracting the 'dependent' and 'independent_variable' from the dataset
x = mydataset.iloc[ : , 0:1]
y = mydataset.iloc[ : , 1:]

# splitting the variables into training_set and testing_set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state = 0)
```

```python
# checking the distribution of the data
plt.scatter(x_train, y_train, color = 'g')
plt.xlabel('Year of experience')
plt.ylabel('Salary')
plt.suptitle('Training dataset')

# creating the linear_regression_model_object
myregobj = linear_model.LinearRegression()
myregobj

# training the model with the training data
myregobj.fit(x_train, y_train)

# drawing the regression_line
plt.plot(x_train, myregobj.predict(x_train), color='r')

myregobj.predict([[7.9]])
```

```python
myregobj.predict(x_test)
plt.scatter(x_test, y_test, color = 'r')
plt.plot(x_test, myregobj.predict(x_test))
plt.xlabel('Years of experience (X_test)')
plt.ylabel('Salary (Y_test)')
plt.suptitle('Testing dataset')
```

```python
# checking the score of the model

myregobj.score(x_train, y_train)
myregobj.score(x_test, y_test)
```
