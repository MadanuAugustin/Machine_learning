```python
# simple linear regression model
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

# loading the test_data
mydata = pd.read_csv('Home_prices.csv')
mydata

# checking the distribution of the data
plt.scatter(mydata.area, mydata.prices, color = 'magenta', marker='+')


# creating the linear regression object
myobject = linear_model.LinearRegression()
myobject



# training the model with the given data
myobject.fit(mydata[['area']], mydata[['prices']])

plt.plot(mydata.area, myobject.predict(mydata[['area']]), color ='limegreen')

# predicting
myobject.predict([[3800]])
```

```python
# testing the model using the test_data
mydf = pd.read_csv('myarea.csv')

predicted = myobject.predict(mydf[['area']])

mydf['my_prices'] = predicted

mydf.to_csv('myarea.csv', index=False)
```
