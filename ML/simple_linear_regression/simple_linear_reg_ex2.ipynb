```python
# simple linear_regression_example_2

# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

# loading the dataset
mydataset = pd.read_csv('ads_sales.csv')
mydataset

# checking the distribution of the data
plt.scatter(mydataset.advertisement, mydataset.sales, color='hotpink')

# creating the linear_regression object
myregobj = linear_model.LinearRegression()
myregobj

# training the model with the given data
myregobj.fit(mydataset[['advertisement']], mydataset[['sales']])

plt.plot(mydataset.advertisement, myregobj.predict(mydataset[['advertisement']]), color = 'g')

myregobj.predict([[130]])
```
