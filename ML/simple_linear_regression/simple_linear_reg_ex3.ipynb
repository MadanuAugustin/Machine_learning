```python
# simple_linear_regression_example_3
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

# loading the data
mydataset = pd.read_csv('canada_per_capita_income.csv')
mydataset

# checking the distribution of the data
plt.scatter(mydataset[['year']], mydataset[['per capita income (US$)']])
plt.xlabel('year')
plt.ylabel('per capita income (US$)')

# creating the linear_regression_object
myregobj = linear_model.LinearRegression()
myregobj


# training the model with the given data
myregobj.fit(mydataset[['year']], mydataset[['per capita income (US$)']])

plt.plot(mydataset[['year']], myregobj.predict(mydataset[['per capita income (US$)']]), color = 'g')

# predicting
myregobj.predict([[2020]])
```
